```python
results = {}

for data in datasets:

    
    bdxo = slimmer_bdxo[(slimmer_bdxo['dataset'] == data)
                        & (slimmer_bdxo['generation'] == 2000)] \
    [['algo', 'training_fitness', 'test_fitness',  'elite_size']].groupby(['algo']) \
    .median().sort_values('test_fitness').reset_index().iloc[0].values.tolist()

    bdxo_results = {'train' : bdxo[1], 'test' : bdxo[2], 'size' : bdxo[3]}

    bdxo_full = slimmer_bdxo[(slimmer_bdxo['dataset'] == data)
                        & (slimmer_bdxo['algo'] == bdxo[0])] \
    [['training_fitness', 'test_fitness',  'elite_size']].values

    dxo = slimmer_dxo[(slimmer_dxo['dataset'] == data)
                        & (slimmer_dxo['generation'] == 2000)] \
    [['algo', 'training_fitness', 'test_fitness',  'elite_size']].groupby(['algo']) \
    .median().sort_values('test_fitness').reset_index().iloc[0].values.tolist()

    dxo_results = {'train' : dxo[1], 'test' : dxo[2], 'size' : dxo[3]}

    dxo_full = slimmer_dxo[(slimmer_dxo['dataset'] == data)
                        & (slimmer_dxo['algo'] == dxo[0])] \
    [['training_fitness', 'test_fitness',  'elite_size']].values

    
    s = slim[data]

    s_results = s[ (s['generation'] == 2000)] \
        [['algo', 'training_fitness', 'test_fitness',  'elite_size']].groupby(['algo']).median() \
    .sort_values('test_fitness').reset_index().iloc[0].values.tolist()

    s_results_d =  {'train' : s_results[1], 'test' : s_results[2], 'size' : s_results[3]}

    s_full = s[ (s['algo'] == s_results[0])] \
    [['training_fitness', 'test_fitness',  'elite_size']].values


    train_p = mannwhitneyu(s_full[:,0], bdxo_full[:,0])[1]
    if train_p > 0.05:
        sign_train = '='
    else:
        if s_results[1] < bdxo[1]:
            sign_train = '<'
        else:
            sign_train = '>'

    test_p = mannwhitneyu(s_full[:,1], bdxo_full[:,1])[1]
    if test_p > 0.05:
        
        sign_test = '='
    else:
        if s_results[2] < bdxo[2]:
            sign_test = '<'
        else:
            sign_test = '>'

    size_p = mannwhitneyu(s_full[:,2], bdxo_full[:,2])[1]
    if size_p > 0.05:
        sign_size = '='
    else:
        if s_results[3] < bdxo[3]:
            sign_size = '<'
        else:
            sign_size = '>'
    #comparisn bdxo dxo
    train_p_2 = mannwhitneyu(dxo_full[:,0], bdxo_full[:,0])[1]
    if train_p_2 > 0.05:
        sign_train_2 = '='
    else:
        if dxo[1] > bdxo[1]:
            sign_train_2 = '<'
        else:
            sign_train_2 = '>'

    test_p_2 = mannwhitneyu(dxo_full[:,1], bdxo_full[:,1])[1]
    if test_p_2 > 0.05:
        
        sign_test_2 = '='
    else:
        if dxo[2] > bdxo[2]:
            sign_test_2 = '<'
        else:
            sign_test_2 = '>'

    size_p_2 = mannwhitneyu(dxo_full[:,2], bdxo_full[:,2])[1]
    if size_p_2 > 0.05:
        sign_size_2 = '='
    else:
        if dxo[3] > bdxo[3]:
            sign_size_2 = '<'
        else:
            sign_size_2 = '>'

    results[data] = {'train' : {'SLIM' : s_results_d['train'], '':sign_train, 'SLIMMER_BDXO' : bdxo_results['train'], 's':sign_train_2, 'SLIMMER_DXO' : dxo_results['train']},
                    'test' : {'SLIM' : s_results_d['test'], '':sign_test, 'SLIMMER_BDXO' : bdxo_results['test'], 's':sign_test_2, 'SLIMMER_DXO' : dxo_results['test']},
                    'size' : {'SLIM' : s_results_d['size'], '':sign_size, 'SLIMMER_BDXO' : bdxo_results['size'], 's':sign_size_2, 'SLIMMER_DXO' : dxo_results['size']},}
    # {'SLIM' : s_results, 'SLIMMER_BDXO' : bdxo_results}
```

```python
variants = slimmer_bdxo.algo.unique().tolist()

results = []

for data in datasets:
    for variant in variants:
        bdxo = slimmer_bdxo[(slimmer_bdxo['dataset'] == data)
                            & (slimmer_bdxo['generation'] == 2000)
                            & (slimmer_bdxo['algo'] == variant)] \
        [['training_fitness', 'test_fitness', 'elite_size']].median().values.tolist()

        bdxo = [round(el, 3) if isinstance(el, float) else el for el in bdxo ]
        
        if len(bdxo) == 0:
            continue
        
        bdxo_full = slimmer_bdxo[(slimmer_bdxo['dataset'] == data) & (slimmer_bdxo['algo'] == variant)] \
        [['training_fitness', 'test_fitness', 'elite_size']].values

        dxo = slimmer_dxo[(slimmer_dxo['dataset'] == data)
                            & (slimmer_dxo['generation'] == 2000)
                            & (slimmer_dxo['algo'] == variant)] \
        [['training_fitness', 'test_fitness', 'elite_size']].median().values.tolist()

        dxo = [round(el, 3) if isinstance(el, float) else el for el in dxo ]
        
        if len(dxo) == 0:
            continue
        
        dxo_full = slimmer_dxo[(slimmer_dxo['dataset'] == data) & (slimmer_dxo['algo'] == variant)] \
        [['training_fitness', 'test_fitness', 'elite_size']].values
        
        s = slim[data]
        s_results = s[(s['generation'] == 2000) & (s['algo'] == variant)] \
        [['training_fitness', 'test_fitness', 'elite_size']].median().values.tolist()

        s_results = [round(el, 3) if isinstance(el, float) else el for el in s_results ]
        
        if len(s_results) == 0:
            continue
        
        s_full = s[(s['algo'] == variant)] [['training_fitness', 'test_fitness', 'elite_size']].values

        def get_significance(values1, values2):
            p_value = mannwhitneyu(values1, values2)[1]
            if p_value > 0.05:
                return '='
            return '<' if np.median(values1) < np.median(values2) else '>'

        def opposite_get_significance(values1, values2):
            p_value = mannwhitneyu(values1, values2)[1]
            if p_value > 0.05:
                return '='
            return '>' if np.median(values1) < np.median(values2) else '<'

        results.append({
            'dataset': data,
            'variant': variant,
            
            'train_SLIM': s_results[0],
            'train_sign': get_significance(s_full[:, 0], bdxo_full[:, 0]),
            'train_SLIMMER_XOBDn': bdxo[0],
            'train_sign_2': opposite_get_significance(dxo_full[:, 0], bdxo_full[:, 0]),
            'train_SLIMMER_XODn': dxo[0],
            
            'test_SLIM': s_results[1],
            'test_sign': get_significance(s_full[:, 1], bdxo_full[:, 1]),
            'test_SLIMMER_XOBDn': bdxo[1],
            'test_sign_2': opposite_get_significance(dxo_full[:, 1], bdxo_full[:, 1]),
            'test_SLIMMER_XODn': dxo[1],
            
            'size_SLIM': s_results[2],
            'size_sign': get_significance(s_full[:, 2], bdxo_full[:, 2]),
            'size_SLIMMER_XOBDn': bdxo[2],
            'size_sign_2': opposite_get_significance(dxo_full[:, 2], bdxo_full[:, 2]),
            'size_SLIMMER_XODn': dxo[2]
        })
```

```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu

algorithms = ['SLIM', 'SLIMMER_XOBDn', 'SLIMMER_XODn']
metrics = ['train', 'test', 'size']

for data in datasets:
    for variant in slimmer_bdxo.algo.unique().tolist():
        # Fetch data for each approach
        bdxo_full = slimmer_bdxo[
            (slimmer_bdxo['dataset'] == data) & 
            (slimmer_bdxo['algo'] == variant)
        ][['training_fitness', 'test_fitness', 'elite_size']].values
        if bdxo_full.shape[0] == 0: continue

        dxo_full = slimmer_dxo[
            (slimmer_dxo['dataset'] == data) & 
            (slimmer_dxo['algo'] == variant)
        ][['training_fitness', 'test_fitness', 'elite_size']].values
        if dxo_full.shape[0] == 0: continue

        s = slim[data]
        s_full = s[
            (s['algo'] == variant)
        ][['training_fitness', 'test_fitness', 'elite_size']].values
        if s_full.shape[0] == 0: continue

        for metric in metrics:
            pval_table = pd.DataFrame(np.nan, index=algorithms, columns=algorithms)

            if metric == 'train':
                pval_table.loc['SLIM', 'SLIMMER_XOBDn'] = mannwhitneyu(s_full[:, 0], bdxo_full[:, 0])[1]
                pval_table.loc['SLIM', 'SLIMMER_XODn'] = mannwhitneyu(s_full[:, 0], dxo_full[:, 0])[1]
                pval_table.loc['SLIMMER_XOBDn', 'SLIMMER_XODn'] = mannwhitneyu(bdxo_full[:, 0], dxo_full[:, 0])[1]

            elif metric == 'test':
                pval_table.loc['SLIM', 'SLIMMER_XOBDn'] = mannwhitneyu(s_full[:, 1], bdxo_full[:, 1])[1]
                pval_table.loc['SLIM', 'SLIMMER_XODn'] = mannwhitneyu(s_full[:, 1], dxo_full[:, 1])[1]
                pval_table.loc['SLIMMER_XOBDn', 'SLIMMER_XODn'] = mannwhitneyu(bdxo_full[:, 1], dxo_full[:, 1])[1]

            elif metric == 'size':
                pval_table.loc['SLIM', 'SLIMMER_XOBDn'] = mannwhitneyu(s_full[:, 2], bdxo_full[:, 2])[1]
                pval_table.loc['SLIM', 'SLIMMER_XODn'] = mannwhitneyu(s_full[:, 2], dxo_full[:, 2])[1]
                pval_table.loc['SLIMMER_XOBDn', 'SLIMMER_XODn'] = mannwhitneyu(bdxo_full[:, 2], dxo_full[:, 2])[1]

            # Print LaTeX format
            print(pval_table.round(4).fillna('--').to_latex(
                caption=f'p-values for {metric} on dataset "{data}" with variant "{variant}"',
                label=f'tab:pvals_{data}_{variant}_{metric}'
            ))
```
